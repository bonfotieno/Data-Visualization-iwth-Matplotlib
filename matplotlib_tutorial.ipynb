# Matplotlib
## 1. Introduction
matplotlib is a 2D plotting library that is used for data visualization

```python
import matplotlib.pyplot as plt
%matplotlib inline 
#the magic above - so that it can render the charts correctly
```

```python
x = [1,2,3,4,5,6,7]
y = [23,56,42,48,43,50,24]
#to plot this, we do:
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y,color='red',linewidth=3,linestyle="dashdot")
```

## 2. Format-Strings in plot function

```python
# we initialized x and y earlier
# check for a list of format strings from the documentation
plt.plot(x,y, 'g+--') #g means green + means plus marker and -- for the dashed line
```

###### if you don't want to use format srings you use keyword arguments

```python
plt.plot(x,y,color='#000040',marker='D',markersize=10,linestyle="",alpha=0.5) #alpha controls transparency
```

## 3. Axes labels, Legend and Grid properties of any 2D plot

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
#in the same chart you can plot several lines
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
# creating the legend by setting labels
plt.plot(days,max_t, label="Max")
plt.plot(days,min_t, label="Min")
plt.plot(days,avg_t, label="Avg")

# setting legend
plt.legend(loc="lower left",fontsize="small") # use loc="best" for the algorithm to identify the suitable position

# Setting grid
plt.grid()
```

## 4. Bar Charts

```python
company = ['GOOGL','AMZN','MSFT','FB']
revenew = [90,136,89,27]
```

## 5. Histograms
Are bar charts where the y-axis contains the frequency of a variable.<br>
we have a dataset of blood sugar level of patients, frequency distribution table is:<br>
80-100    -> Normal <br>
100-125   -> Pre-diabetic <br>
125 Above -> Diabetic <br><br>
Questions:<br>
1) how many patients are normal? <br>
2) how many of them are pre-diabetic? <br>
3) how many are diabetic? <br>
The dataset is as bellow:





```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 80, 77, 82, 129]
#when working with histogram we only need single dimension arrays and the y-axis will contain the frequency of those different numbers
```

```python
plt.hist(blood_sugar)
#by default it plots 10 bins(it's a range)-the bars represent them
```

```python
plt.hist(blood_sugar,bins=3, rwidth=0.8) # rwidth is the distance between the bars 
```

##### customizing the range of each bin 
80-100 -> Normal <br>
100-125 -> Pre-diabetic <br>
125 Above -> Diabetic

```python
plt.hist(blood_sugar, color='blue', histtype='step', #change the histogram type to step
         bins=[80,100,125,150], # We have customized the range of each bin.150 is just a maximum value
         rwidth=0.8) 
```

##### if you have two different samples - Ploting two histogram side by side

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 80, 77, 82, 129]
blood_sugar_women = [67, 93, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.xlabel('sugar range')
plt.ylabel('Total Number Of Patients')
plt.title("Blood Sugar Analysis")
plt.hist(
    [blood_sugar_men,blood_sugar_women],color=['blue','lightblue'], # we use two different colors for that of men and women 
    bins=[80,100,125,150],
    rwidth=0.95,
    label=['men','women'],
    orientation='horizontal'
)
plt.legend()
```

## 6.Pie Charts 

```python
expense_vals = [1400,600,200,410,250]
expense_labels = ['home rent','Food','Phone/Internet Bill','Car','Other Utilities']
```

```python
plt.axis("equal")
plt.pie(expense_vals,
        labels = expense_labels,
        autopct="%0.2f%%",       # to see the percentages, 0.2 specifies decimal points, 0.0 for no decimal points
        explode=[0,0,0.4,0,0],   # take out pies, if you specify bigger value the pie explodes further
        startangle=45,           # the chart starts at 45 degrees from right
        radius=1.5,
        shadow = True
       )
plt.show() # shows the pie chart only
```
